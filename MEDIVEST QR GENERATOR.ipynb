```python
import pandas as pd

asset_list = pd.read_csv('./Asset List_SBN.csv')
```

```python
clean_asset_list = asset_list[asset_list['Asset Status'] == 'Active'].drop(
    [
        'Company Name',
        'State', 
        'Hospital Name', 
        'Model', 
        'Serial No', 
        'Asset Type Description', 
        'Asset Type Code',
        'Asset Status',
        'User Area Name',
        'User Department Name',
        'User Department Code',
        'User Area Code',
        'Manufacturer',
        'User Location Code',
        'User Location Name',
        'Real Time Status',
        'POC Flag'
    ],axis=1)
asset_to_track = clean_asset_list[pd.notnull(clean_asset_list['Total Maintenance Activity'])] \
    .reset_index().drop('index',axis=1)
activity = asset_to_track.replace(to_replace='No', value= 0).sum()['Total Maintenance Activity']
print("Asset To Track with QR Code = {:,.0f}".format(len(asset_to_track)))
print("Total Activity QR Code = {0:,.0f}".format(int(activity)))
print("-----------------------------------")
print("Total QR Code To Generate = {:,.0f}".format(int(len(asset_to_track)+activity)))

gro = asset_to_track.groupby('Asset No.')
new_csv = asset_to_track.replace(to_replace='No', value= '0')

new_csv
```

```python
class Formatting:
    def asset_code(self, org_code, asset_no, id):
        org = str(org_code).zfill(4)
        asset = str(asset_no)
        no_id = str(id).zfill(8)
        return org + '-' + asset + '-' + no_id

fmt = Formatting()
activity_columns = list(new_csv.columns)[1:-1]

asset_txt = []
for key,i in enumerate(new_csv.to_dict('records')[:5]):
    asset = {}
    asset['asset_no'] = i['Asset No.']
    asset['activity'] = []
    for a in activity_columns:
        if i[a] == '1':
            activity = {
                "name": a,
                "code": fmt.asset_code(1,i['Asset No.'],key)
            }
            asset['activity'].append(activity)
    asset_txt.append(asset)
    
```
